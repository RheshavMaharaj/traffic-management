##Linear Reg for Hourly Sample

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import sys

def main():
    hourly_samples_path = 'https://drive.google.com/uc?id=1URz3Zp1sS9n8cK0H47RJiNdV55LxXiYY'
    df = pd.read_csv(hourly_samples_path)

    df.groupby(['station_key']) #group the data by station
    
    x, y = pre_processing(df)
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
    
    model = LinearRegression()
    model.fit(x_train,y_train)

    evaluate(model, x_train, y_train, x_test, y_test)

    return

def pre_processing(dataframe):
    features = dataframe.iloc[0:20, 10:39]
    columns_drop = ["month","day_of_week", "public_holiday", "school_holiday"]
    features = features.drop(columns_drop, axis=1)
    target = dataframe.iloc[0:20, 4]
    
    return features, target

def evaluate(model, x_train, y_train, x_test, y_test ):
    if(model == None):
        print("Model does not exist")
        return

    print(f"Train accuracy {model.score(x_train,y_train)} ")
    print(f"Test accuracy {model.score(x_test,y_test)} ")
    return

main()
```
