```python

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, KFold, cross_val_score

import sys

from sklearn.metrics import accuracy_score
```

```python

#df = 'https://drive.google.com/uc?export=download&id=19DIL99UGIzhPZQ_VDuyrDmGPvy5VMGAn'

df = 'https://drive.google.com/uc?export=download&id=1CE5IoJSha20kZE8H5gT9XRo33U0Fzub8'
raw_dataset = pd.read_csv(df)
```

```python
dataset = raw_dataset.copy()
pd.set_option("display.max_columns", None)
dataset.head()
```

```python
def pre_processing(dataframe):
    features = dataframe.iloc[:, 0:32]
    #columns_drop = ['station_key', 'day_of_week', 'public_holiday', 'school_holiday', 'full_name', 'latitude', 'longitude', 'daily_total']
    columns_drop = ['station_key', 'day_of_week', 'school_holiday', 'full_name', 'wgs84_latitude', 'wgs84_longitude', 'daily_total']
    features = features.drop(columns_drop, axis=1)
    dataframe = dataset.dropna()
    target = dataframe.iloc[:, 32]
    
    return features, target
```

```python
def evaluate(model, x_train, y_train, x_test, y_test ):
    if(model == None):
        print("Model does not exist")
        return

    print(f"Train accuracy {model.score(x_train,y_train)} ")
    print(f"Test accuracy {model.score(x_test,y_test)} ")
    return 
```

```python
def cross_validation(dataset, model, x, y):
  k = 5
  kf = KFold(n_splits=k, random_state=None)
  
  result = cross_val_score(model , x, y, cv = kf)
  print("Avg accuracy: {}".format(result.mean()))
```

```python
x, y = pre_processing(dataset)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)

model = LinearRegression()
model.fit(x_train.values,y_train.values)

evaluate(model, x_train, y_train, x_test, y_test)

cross_validation(dataset, model, x, y)
```
