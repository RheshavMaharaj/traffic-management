```python
finalDF = pd.merge(hourlyDF, stationDF, 
                   on='station_key', 
                   how='left')
```

```python
pd.set_option("display.max_columns", None)
```

```python
drop_columns1 = finalDF.iloc[:, 0:4]
drop_columns2 = finalDF.iloc[:, 5:11]
drop_columns3 = finalDF.iloc[:, 39:48]
drop_columns4 = finalDF.iloc[:, 49:75]
drop_columns5 = finalDF.iloc[:, 77:82]

finalDF = finalDF.drop(drop_columns1, axis=1)
finalDF = finalDF.drop(drop_columns2, axis=1)
finalDF = finalDF.drop(drop_columns3, axis=1)
finalDF = finalDF.drop(drop_columns4, axis=1)
finalDF = finalDF.drop(drop_columns5, axis=1)

finalDF = finalDF.dropna()

finalDF.tail()
```

```python
finalDF['threshold'] = finalDF[['hour_00', 'hour_01', 'hour_02', 'hour_03', 'hour_04', 'hour_05', 
                                'hour_06', 'hour_07', 'hour_08', 'hour_09', 'hour_10', 'hour_11', 
                                'hour_12', 'hour_13', 'hour_14', 'hour_15', 'hour_16', 'hour_17', 
                                'hour_18', 'hour_19', 'hour_20', 'hour_21', 'hour_22', 'hour_23']].max(axis=1)

finalDF.loc[:,'threshold'] *= 0.8
```

```python
finalDF.tail()
```

```python
finalDF.drop(finalDF.index[finalDF['daily_total'] < 500], inplace = True)
```

```python
finalDF.tail()
```

```python
finalDF = finalDF.dropna()
finalDF.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')
```

```python
finalDF.tail()
```

```python
cols = ["public_holiday","school_holiday"]
finalDF[cols] = finalDF[cols].replace({'0':False, '1':True, 0:False, 1:True})
```
